#                                                   Feature Extracting :)

we will make new columns that help answer business questions

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("cleaned_supply_chain.csv",encoding='latin1')
df.head()
```

```python
df[['order date (DateOrders)', 'shipping date (DateOrders)']].dtypes
#we have make them in datetime dtype
```

```python
df['order date (DateOrders)'] = pd.to_datetime(
    df['order date (DateOrders)'],
    errors='coerce'
)

df['shipping date (DateOrders)'] = pd.to_datetime(
    df['shipping date (DateOrders)'],
    errors='coerce'
)
```

```python
df[['order date (DateOrders)', 'shipping date (DateOrders)']].dtypes
```

```python
df['Order Year'] = df['order date (DateOrders)'].dt.year
```

```python
df['Order Month'] = df['order date (DateOrders)'].dt.month_name()
```

```python
df['Order Quarter'] = df['order date (DateOrders)'].dt.quarter
```

```python
df['Order Day'] = df['order date (DateOrders)'].dt.day_name()
```

```python
df.head()
```

# checking things of shipping time

```python
df['Shipping Time'] = (
    df['shipping date (DateOrders)'] -
    df['order date (DateOrders)']
).dt.days
```

```python
df[['Shipping Time','Days for shipping (real)']].head()
```

```python
#everything is good
```

```python
#shipping delay
df['Shipping Delay'] = (
    df['Days for shipping (real)'] -
    df['Days for shipment (scheduled)']
)
df['Shipping Delay'].head(5)
# <0 == earlier dilevery 
# =0 == on time dilevary
# >0 == delayed dilvery
```

```python
#profit margin
df['Profit Margin (%)'] = (
    df['Order Profit Per Order'] / df['Sales']
) * 100
```

```python
df['Profit Margin (%)'].head(7)
```

```python
#high value orders
df['High Value Order'] = np.where(
    df['Sales'] >= df['Sales'].median(),
    'High',
    'Low'
)
df['High Value Order'].head(7)
```

```python
#discount
df['Calculated Discount'] = (
    df['Order Item Product Price'] *
    df['Order Item Quantity'] *
    df['Order Item Discount Rate']
)
df['Calculated Discount'].head(7)
```

```python
#weekend order
df['Weekend Order'] = np.where(
    df['Order Day'].isin(['Saturday', 'Sunday']),
    'Yes',
    'No'
)
```

```python
df['Weekend Order'].head()
```

# Saving the data

```python
df.to_csv('Feature_extracted_data.csv', index=False)
```
